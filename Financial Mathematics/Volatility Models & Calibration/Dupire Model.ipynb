### Derivation of Dupire's Formula

We first introduce a discounting factor $P(t, T)$ as following:
$$P(t, T) = \exp(-\int^T_t r_s ds)$$
meaning that all the discount factor solely depends on the instant risk-free rate throught time (0, T)

Assuming the asset price at time t can be determined by a density function $f=f(S,t)$, the pricing of an European call option is then given as
$$C(S_t, K) = P(t, T)E \left[(S_T - K)^{+} \right]
            = P(t, T)\int^\infty_K (S_T - K)f(S_T, T)dS$$
Here, the function P represents the discount factor, the $S_T-K$ refers to the expected return for a Vanilla call option, and function f refers to the distribution of the asset price at maturity.

Before deducing the time derivative of the asset price $C(S_t, K)$, we shall give a description of the movement of the asset price as following
$$dS_t = \mu (S_t, t)dt + \sigma(S_t, t)dW_t$$ 
Here, the movement is described by an Ito Process, where the drift rate and the volatility are not constant

According to Kolmogorov forward equation, for a given Ito Process, the time derivative of the asset price is expressed as 
$$\frac{\partial f}{\partial t} = -\frac{\partial}{\partial S}\left[\mu(S,t)f(S,t)\right] + \frac{1}{2}\frac{\partial^2}{\partial S^2}\left[\sigma ^2(S,t)f(S,t)\right]$$

Bringing the Ito Process into the forward equation, we get
$$\frac{\partial f}{\partial t} = -\frac{\partial}{\partial S}\left[(r_t - q_t)Sf(S,t)\right] + \frac{1}{2}\frac{\partial^2}{\partial S^2}\left[(\sigma_L S)^2f(S,t)\right] \, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(1)$$

Then, we take the time derivative of the asset price $C(S_t, K)$
$$\frac{\partial C}{\partial T} = \frac{\partial}{\partial T}\left[P(t, T)\int_K^\infty(S_T-K)f(S_T,T)dS\right]$$
By Chain Rule, we have
$$\frac{\partial C}{\partial T} = \left[\frac{\partial}{\partial T}P(t, T)\right]\int_K^\infty(S_T-K)f(S_T,T)dS + P(t, T)\left[\frac{\partial}{\partial T}\int_K^\infty(S_T-K)f(S_T,T)dS\right]$$
Taking the time derivative of $P(t, T)$,
$$\frac{\partial}{\partial T}P(t, T) = \frac{\partial}{\partial T} \exp(-\int^T_t r_s ds)
= -r_T \exp(-\int^T_t r_s ds)$$
Thus, $$\frac{\partial C}{\partial T} = -r_T \left[ \exp(-\int^T_t r_s ds)\int_K^\infty(S_T-K)f(S_T,T)dS \right] + P(t, T)\left[\frac{\partial}{\partial T}\int_K^\infty(S_T-K)f(S_T,T)dS\right]$$
$$=-r_TC(S_t, K) + P(t, T)\left[\int_K^\infty(S_T-K)\frac{\partial}{\partial T}f(S_T,T)dS\right]\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(2)$$
Notice that the time derivative in the second term is now applied to the density function inside the integral, as $(S_T - K)$ remains constant in differentiation.

Take the time derivative of asset prices in (1) into (2),
$$\frac{\partial C}{\partial T} =-r_TC(S_t, K) + P(t, T) \Bigg[ \int_K^\infty(S_T-K) \left\{ -\frac{\partial}{\partial S}\left[(r_t - q_t)Sf(S,t)\right] + \frac{1}{2}\frac{\partial^2}{\partial S^2}\left[(\sigma_L S)^2f(S,t)\right]\right\}dS \Bigg]\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(3)$$

Let $$I_1 = (r_T - q_T)\int^{\infty}_K(S_T - K)\frac{\partial}{\partial S}\left[Sf(S,T)\right]dS$$
$$I_2 = \int^{\infty}_K(S_T - K)\frac{\partial^2}{\partial S^2}\left[\sigma_L^2S^2f(S,T)\right]dS$$

Integrating $I_1, I_2$ by parts, we get
$$I_1 = -(r_T - q_T)\big(\frac{C(S_t, K)}{P(t, T)} - \frac{K}{P(t, T)}\frac{\partial C}{\partial K}\big)$$
$$I_2 = \sigma^2_2K^2\frac{\partial^2 C / \partial K^2}{P(t, T)}$$

Take $I_1, I_2$ back into the equation,
$$\frac{\partial C}{\partial T} = -r_TC(S_t, K) + P(t, T)\left[-I_1 + \frac{1}{2}I_2\right]$$
$$\frac{\partial C}{\partial T} =-r_TC(S_t, K) + P(t, T)(r_T - q_T)\big(\frac{C(S_t, K)}{P(t, T)} - \frac{K}{P(t, T)}\frac{\partial C}{\partial K}\big) + \frac{1}{2}P(t,T)\sigma^2_LK^2\frac{\partial^2 C / \partial K^2}{P(t, T)}$$
$$\frac{\partial C}{\partial T} = -r_TC(S_t, K) + (r_T - q_T)\big(C(S_t, K) - K\frac{\partial C}{\partial K}\big) + \frac{1}{2}\sigma^2_LK^2\frac{\partial^2 C }{\partial K^2}\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(4)$$

Here, the first term $-r_TC$ represents the cost of holding the position, discounted at the risk-free rate, and the second term $(r_T - q_T)(C - K\frac{\partial C}{\partial K})$ signifies the drift term, or the expected growth due to the accruing interest. 

The combines effects of the risk-free rate $r_T$ and the dividend yield $q_T$ are modeled and $(C - K\frac{\partial C}{\partial K})$ is made to represent the portfolio that holds an option and shorts a fraction $K$ of the underlying asset (essentially a delta-hedging term).

By transforming the equation, we get the Dupire Local Volatility
$$\sigma^2_L(K,T) = \frac{\frac{\partial C}{\partial T} + r_TC -(r_T-q_T)(C-K\frac{\partial C}{\partial K})}{\frac{1}{2}K^2\frac{\partial^2 C }{\partial K^2}}$$

Then, were we to express the local volatility as a function of the forward price $F_T = S_0 \exp \left({\int^T_0 \mu(t)dt }\right)$ instead of the spot price, we shall remove the drift terms from eq. (4). That is, 
$$\frac{\partial C}{\partial T} = \frac{1}{2}\sigma^2K^2\frac{\partial^2C}{\partial K^2}$$

This gives
$$\sigma^2(K,T,S_0) = \frac{\frac{\partial C}{\partial T}}{\frac{1}{2}K^2\frac{\partial^2 C }{\partial K^2}}\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(5)$$

## Express Local Volatility by Implied Volatility

In practical scenarios, it is much more computationally efficient to model the local volatility in terms of the total variance $\omega = \sigma^2_{BS}T$ indicated by the Black-Scholes model, as it merely incorporates two dimensional variables 
$$\omega(S_0, K,T):=\sigma^2_{BS}(S_0, K,T)T$$

Meanwhile, we define the log-strike, or the moneyness of the strike price with regard to the forward price of the option
$$y=log(\frac{K}{F_T})\,\,\,\,\, F_T = S_0 \exp \left({\int^T_0 \mu(t)dt }\right)$$

From the BS model, we have the pricing for an European call option as
$$P_{call} = S_0N(d_1)-Ke^{-rT}N(d_2)$$
where $d_1 = \frac{log(S_0/K)+(r+\sigma^2/s)T}{\sigma \sqrt{T}}$, $d_2 = d_1 - \sigma \sqrt{T}$

Make transformation to the log-strike, we have $e^y = \frac{K}{F_T}$, or $K = e^yF_T$.

Substituting $Ke^{-rT} = e^yF_Te^{-rT}$ and $S_0 = \frac{F_T}{e^{rT}}$,
$$P_{call} = \frac{F_T}{e^{rT}}N(d_1)-e^yF_Te^{-rT}N(d_2) = F_T \{ N(d_1)-e^yN(d_2) \}\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(6)$$

Meanwhile, we need to express $ln(S_0/K)$. 

Since $y = log(K/F_T)$, 
$log(K/S_0) = y - log(S_0/F_T) = y - log(\frac{\frac{F_T}{e^{rT}}}{F_T}) = y + log(e^{rT}) = y + rT$
Substitute these into the expressions for $d_1$ and $d_2$,
$$d_1 = \frac{-(y+rT) + (r+\sigma^2/2)T}{\sigma \sqrt{T}} = \frac{-y+\sigma^2T/2}{\sigma \sqrt{T}} = \frac{-y}{\sqrt{\omega}} + \frac{\sqrt{\omega}}{2}$$
$$d_2 = \frac{-y}{\sqrt{\omega}} - \frac{\sqrt{\omega}}{2}$$

To express local volatility in terms of implied volatility, we need the three derivatives $\frac{\partial C}{\partial T}$, $\frac{\partial C}{\partial K}$, and $\frac{\partial^2 C}{\partial K^2}$ in the form of y.

Recall that $C(S_0, K, T) = C(S_0, F_Te^y, T)$.

Then, we have by the chain rule
$$\frac{\partial C}{\partial y} = \frac{\partial C}{\partial K}\frac{\partial K}{\partial y} = \frac{\partial C}{\partial K} K$$
$$\frac{\partial^2 C}{\partial y^2} = \frac{\partial}{\partial y}(\frac{\partial C}{\partial K})K + \frac{\partial C}{\partial K}\frac{\partial K}{\partial y} = \frac{\partial^2 C}{\partial K^2}K^2 + \frac{\partial C}{\partial y}$$
and $$\frac{\partial C}{\partial T} = \frac{\partial C}{\partial T} + \frac{\partial C}{\partial y}\mu_T$$
since $K = S_0 \exp(\int^T_0 \mu_t dt +y)$ so that $\frac{\partial K}{\partial T} = K \mu_T$

Taking these into the Dupire equation (4),
$$\frac{\partial C}{\partial T} = \frac{v_L}{2} \left[ \frac{\partial^2 C}{\partial y^2} - \frac{\partial C}{\partial y} \right] + \mu_TC\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(6.2)$$
where $v_L = \sigma^2(K,T)$ is the local variance.

Now, we take derivatives of the BS model.

Notice that $N'(x) = n(x)x'$ and $n'(x) = -xn(x)x'$ for the derivative of the standard normal cdf and pdf.

So, $n(d_1) = \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}(d_2 + \sqrt{\omega})^2} = \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}(d_2^2 + 2d_2\sqrt{\omega}+\omega)} = n(d_2)e^{-d_2\sqrt{\omega} - \frac{1}{2}\omega} = n(d_2)e^y$

Then, from BS eq. (6), the first derivative with respect to $\omega$ is 
$$\frac{\partial C_{BS}}{\partial \omega} = F_T\left[n(d_1)d_{1\omega} - e^yn(d_2)d_{2\omega}\right]$$
$$=F_T\left[n(d_2)e^y(d_{2\omega} + \frac{1}{2}\omega^{-\frac{1}{2}}) - e^yn(d_2)d_{2\omega}\right]$$
$$=\frac{1}{2}F_Te^y \left[n(d_2)\omega ^{-\frac{1}{2}} \right]$$

where $d_{1\omega}$ is the first derivative of $d_1$ with respect to $\omega$

Further, we take the second derivative
$$\frac{\partial^2 C_{BS}}{\partial \omega^2} = \frac{\partial C_{BS}}{\partial \omega}\left[-\frac{1}{8} - \frac{1}{2\omega} + \frac{y^2}{2\omega^2}\right]\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(7)$$

We now have the first identity eq. (7) for later term replacement. Then, the second identity is
$$\frac{\partial^2 C_{BS}}{\partial \omega \partial y} = \frac{1}{2}F_T\omega^{-\frac{1}{2}}\frac{\partial}{\partial y}\left[e^y n(d_2)\right] = \frac{\partial C_{BS}}{\partial \omega}(\frac{1}{2} - \frac{y}{\omega})\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(8)$$

Finally, the third identity is given as
$$\frac{\partial C_{BS}}{\partial y} = F_T\left[n(d_1)d_{1y} - e^y N(d_2) - e^yn(d_2)d_{2y}\right] = -F_Te^yN(d_2)$$

$$\frac{\partial^2 C_{BS}}{\partial y^2} = \frac{\partial C_{BS}}{\partial y} + 2\frac{\partial C_{BS}}{\partial \omega}\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(9)$$

Knowing that the total volatility $\omega$ depends on $K$ and $K$ depends on $y$, $K=F_Te^y$, we can write $\omega = \omega(y)$ and 
$$C(S_0, K, T) = C_{BS}(S_0, F_Te^y, \omega(y), T)$$

We then need the derivatives of the market call price $C$ in terms of the Black- Scholes call price $C_{BS}$. The first derivative we need is 
$$\frac{\partial C}{\partial y} = \frac{\partial C_{BS}}{\partial y} + \frac{\partial C_{BS}}{\partial \omega}\frac{\partial \omega}{\partial y}$$
$$=a(\omega, y) + b(\omega, y)c(y)\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(10)$$
where the first term refers to components of $C_{BS}$ independent of y and the second arises from those dependent on $\omega$ or $y$.

The second derivative is 
$$\frac{\partial^2 C}{\partial y^2} = \frac{\partial^2 C_{BS}}{\partial y^2} + 2\frac{\partial^2 C_{BS}}{\partial y\partial \omega}\frac{\partial \omega}{\partial y} + \frac{\partial C_{BS}}{\partial \omega}\frac{\partial^2 \omega}{\partial y^2} + \frac{\partial ^2 C_{BS}}{\partial \omega^2} (\frac{\partial \omega}{\partial y})^2\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(11)$$

Similarly, the third one is
$$\frac{\partial C}{\partial T} = \frac{\partial C_{BS}}{\partial T} + \frac{\partial C_{BS}}{\partial \omega}\frac{\partial \omega}{\partial T} = \mu_TC + \frac{\partial C_{BS}}{\partial \omega}\frac{\partial \omega}{\partial T}\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(12)$$

The second equality follows because the only explicit dependence of $C_{BS}$ on $T$ is through the forward price $F_T$ , even though $C_{BS}$ depends implicitly on $T$ through $y$ and $\omega$.

Finally, we substitute partial derivatives in (10), (11), and (12) into the equation (6.2), while removing the term $\mu_TC$.
As a reminder,
$$\frac{\partial C}{\partial T} = \frac{v_L}{2} \left[ \frac{\partial^2 C}{\partial y^2} - \frac{\partial C}{\partial y} \right] + \mu_TC\, \, \, \, \,\, \, \, \, \,\,\,\,\,\,\,(6.2)$$

So, 
$$\frac{\partial C_{BS}}{\partial \omega}\frac{\partial \omega}{\partial T} = \frac{v_L}{2}\left[ \frac{\partial^2 C_{BS}}{\partial y^2} + 2\frac{\partial^2 C_{BS}}{\partial y\partial \omega}\frac{\partial \omega}{\partial y} + \frac{\partial C_{BS}}{\partial \omega}\frac{\partial^2 \omega}{\partial y^2} + \frac{\partial ^2 C_{BS}}{\partial \omega^2} (\frac{\partial \omega}{\partial y})^2 - \frac{\partial C_{BS}}{\partial y} + \frac{\partial C_{BS}}{\partial \omega}\frac{\partial \omega}{\partial y}\right]$$

Then, we substitute for $\frac{\partial^2 C_{BS}}{\partial \omega^2}$, $\frac{\partial^2 C_{BS}}{\partial \omega \partial y}$, and $\frac{\partial^2 C_{BS}}{\partial y^2}$ from identities in (7), (8), and (9)

$$\frac{\partial C_{BS}}{\partial \omega}\frac{\partial \omega}{\partial T} = \frac{v_L}{2}\frac{\partial C_{BS}}{\partial \omega} \left[2 + 2(\frac{1}{2} - \frac{y}{\omega})\frac{\partial \omega}{\partial y} + (-\frac{1}{8} - \frac{1}{2\omega} + \frac{y^2}{2\omega^2})(\frac{\partial \omega}{\partial y})^2 + \frac{\partial^2 \omega}{\partial y^2} - \frac{\partial \omega}{\partial y}\right]$$

Remove the factor $\frac{\partial C_{BS}}{\partial \omega}$ from both sides and simplify to get the formula for local volatility expressed in terms of the implied volatility.
$$v_L = \frac{\partial \omega / \partial T}{1 - \frac{y}{\omega} + \frac{1}{4}\left(-\frac{1}{4} - \frac{1}{\omega} + \frac{y^2}{\omega^2}\right)\left(\frac{\partial \omega}{\partial y}\right)^2 + \frac{1}{2}\frac{\partial^2 \omega}{\partial y^2}}$$

```python
data["w"] = data["market_imp_vol"] ** 2 * data["maturity"]  # Total Variance (0, T)
data["y"] = np.log(data["exerciseprice"] / data["F"])
option = data.loc[data["实虚值"]=="虚值"]  
option = option.sort_values(["contract_month","y"])
option
```

To calculate the partial derivative of $\omega$ w.r.t. time and y using the difference method, we employs three-spline interpolation for exerciseprice and the flat forward for the time, considering a relatively small number of observed samples.

We then define a function for calculating $\omega$

```python
#For each maturity, we have a corresponding cubicsplines of relationship 
#between total variance and the moneyness.
#Alternatively, we are creating the cubic splines of the moneyness-variance
#for each maturity.
def get_spline(data): 
    spline = []
    for m in data["maturity"].unique():
        moneyness = data.loc[data["maturity"] == m]["y"] 
        sample_volatility = data.loc[data["maturity"] == m]["w"]
        cs = CubicSpline(x = moneyness,y = sample_volatility, extrapolate=True)
        spline.append(cs)
    return spline

#Use linear interpolation to get a continuous surface of the 'moneyness-variance'
#curve through time.
def get_total_v(data, spline, y, t):
    total_v = [float(cs(y)) for cs in spline]
    f = interp1d(x = data["maturity"].unique(),
                 y = total_v,
                 kind = "linear",
                 fill_value="extrapolate") 
    vol_spec = float(f(t))
    return vol_spec


spline = get_spline(option)
```

Now, we consider the Dupire's formula.

We then define a function for calculating derivatives with finite-differentiation.
$$\frac{\partial \omega}{\partial t} = \frac{ \omega_{y, t+\Delta t} - \omega_{y, t}}{\Delta t}$$
$$\frac{\partial \omega}{\partial y} = \frac{ \omega_{y+\Delta y, t} - \omega_{y-\Delta y, t}}{2\Delta y}$$
$$\frac{\partial^2 \omega}{\partial y^2} = \frac{ \omega_{y+\Delta y, t} +\omega_{y-\Delta y, t} - 2\omega_{y, t}}{(\Delta y)^2}$$

```python
def get_diff(data, spline, y, t):
    yt = get_total_v(data, spline, y, t)
    t_up = get_total_v(data, spline, y, t*(1+0.001))
    
    y_up = get_total_v(data, spline, y*(1+0.001), t)
    y_down = get_total_v(data, spline, y*(1-0.001), t)
    
    dw_dt = (t_up - yt) / (t*0.001) # Forward Second-Degree
    dw_dy = (y_up - y_down) / (y*0.001*2)
    dw_dy2 = (y_up + y_down - 2*yt) / (y*0.001)**2
    return dw_dt, dw_dy, dw_dy2
```

Then, employ the Dupire equation for estimating local volatility

```python
def get_v(data, spline, y, t, tol=1e-8):
    total_v = get_total_v(data, spline, y, t)
    dw_dt, dw_dy, dw_dy2 = get_diff(data, spline, y, t)
    numetator = dw_dt
    denonimator = (1
                   - y/ total_v * dw_dy
                   + 0.25*(-0.25 - 1 / total_v + y**2 / total_v**2) * (dw_dy**2) 
                   + 0.5*dw_dy2)
    
    local_variance = numetator / denonimator
    
    if local_variance < 0:
        local_variance  = tol
    
    return np.sqrt(local_variance)
```

We adopt the same plotting function from the note 'Volatility'

```python
def plot_ivs(t, k, v):
    fig = plt.figure(figsize=(12, 7))
    ax = plt.axes(projection='3d')
    norm = mpl.colors.Normalize(vmin=0.05, vmax=0.3)
    
    surf = ax.plot_surface(t, k, v, rstride=1, cstride=1,
    cmap=plt.cm.coolwarm,norm = norm, linewidth=0.5, antialiased=True)
    
    ax.set_xlabel('maturity')
    ax.set_ylabel('strike')
    ax.set_zlabel('market_imp_vol')
    ax.set_zlim((0,0.5))
    fig.colorbar(surf, shrink=0.25, aspect=5)
```

```python
def get_v_group(data, spline):
    t_array = np.linspace(data["maturity"].min(), data["maturity"].max(), 15)
    y_array = np.linspace(data["y"].min(), data["y"].max(), 20)   
    t_grid, y_grid = np.meshgrid(t_array, y_array) 
    v = np.zeros_like(y_grid)

    for t_idx, t1 in enumerate(t_array):
        for y_idx, y1 in enumerate(y_array):
            v[y_idx, t_idx] = get_v(data, spline, y1, t1)
    return t_grid, y_grid, v


t, y, v = get_v_group(option, spline)
plot_ivs(t,y,v)
```
